```python
debug = True

def simplex(A, vars):
  #print all rows
  if debug:
    for i in range(0, len(A)): #iterate thru all rows
      print(A[i][:])

  print("STARTING SIMPLEX ON:")
  for i in range(0, len(A)): #iterate thru all rows
    print(A[i][:])
  #for i in range(0, 2): #replace w infinite while loop later
  flag_soln = True
  while (True):
    #look for pivot col
    biggest = 0
    pivot_col_idx = -1
    for i in range(0, len(A[-1])): #iterate thru all cols
      if A[-1][i] < biggest:
        biggest = A[-1][i]
        pivot_col_idx = i
    #check if pivot col found
    if (pivot_col_idx != -1):
      if debug:
        print("pivot col: " + str(A[-1][pivot_col_idx])) #print last row, elem in pivot_col_idx
    else:
      if debug:
        print("pivot col not found, exiting")
      break #leave while loop bc no pivot col found

    #look for pivot row
    smallest_non_neg_ratio = 9999999
    pivot_row_idx = -1
    for i in range(0, len(A)-1): #iterate thru all rows but do not include last row bc is eqn to max
      if (A[i][pivot_col_idx] != 0): #ensure no div by zero
        curr_ratio = A[i][-1] / A[i][pivot_col_idx] #last elem divided by elem of pivot row
        if (curr_ratio < smallest_non_neg_ratio) and (curr_ratio >= 0):
          smallest_non_neg_ratio = curr_ratio
          pivot_row_idx = i
    #check if pivot row found
    if (pivot_row_idx != -1):
      if debug:
        print("pivot row: " + str(A[pivot_row_idx][:])) #print entire pivot row
    else:
      if debug:
        print("pivot row not found, exiting")
      flag_soln = False
      break #leave while loop bc no pivot row found

    #change pivot row to have 1 in pivot column entry
    factor = A[pivot_row_idx][pivot_col_idx]
    for i in range(0, len(A[pivot_row_idx])): #iterate thru all cols
      A[pivot_row_idx][i] = A[pivot_row_idx][i] / factor

    #transform all other row to clear the entry in the pivot column
    for i in range(0, len(A)): #iterate thru all rows
      if (i != pivot_row_idx): #do not apply to pivot row
        factor = A[i][pivot_col_idx] / A[pivot_row_idx][pivot_col_idx]
        for j in range(0, len(A[i])): #iterate thru all cols
          A[i][j] = A[i][j] - (factor * A[pivot_row_idx][j])

    #print all rows
    if (debug):
      for i in range(0, len(A)): #iterate thru all rows
        print(A[i][:])

  #out of loop, display solution
  if not flag_soln:
    print("no solution possible: could not find pivot row with non-negative ratio")
  else:
    #display solution matrix:
    print("final solution is:")
    for i in range(0, len(vars)): #iterate thru all vars
      sum = 0
      flag_neg = False
      one_row_idx = -1
      for j in range(0, len(A)): #iterate thru all rows in that column
        if (A[j][i] < 0):
          flag_neg = True
        if (A[j][i] == 1):
          one_row_idx = j
        sum = sum + A[j][i]
      if not flag_neg and (sum == 1):
        print(str(vars[i]) + " = " + str(A[one_row_idx][-1]))


A_test = [[1, 2, 1, 0, 0, 0, 16], [1,1,0,1,0,0,9], [3,2,0,0,1,0,24], [-40,-30,0,0,0,1,0]]
vars_test = ["x1", "x2", "s1", "s2", "s3", "P", "rhs"]
A_sample = [[1, 0, 0, 1, 0, 0, 0, 0, 30], [0, 1, 0, 0, 1, 0, 0, 0, 50], [0, 0, 1, 0, 0, 1, 0, 0, 100], [1, 1, 1, 0, 0, 0, 1, 0, 140],[-5,-4,-3,0,0,0,0,1,0]]
vars_sample = ["x1", "x2", "x3", "s1", "s2", "s3", "s4", "P", "rhs"]

simplex(A_test, vars_test)
simplex(A_sample, vars_sample)
```
